# 관계 살펴보기 (수치형 vs 수치형)

jointplot(data = df, x = , y = , kind = 'scatter')
- kind 에 따라 그래프의  형태 변경됨
    - 'scatter'(기본값) : point
    - 'reg': point + regression
    - 'kde' : 누적 분포 차트 like 지도

```python
# jointplot을 이용해 두 수치데이터간의 관계와, 각 데이터의 분포를 함께 확인 할 수 있습니다.
# kind 옵션을 통해 여러 타입으로 시각화 할 수 있는데, 기본적으로 kind = 'scatter'으로 지정되어있습니다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill')
```

```python
# kind = 'kde'로 지정할 경우, 밀도함수를 이용해 시각화 가능합니다.
# 진하게 표시될 수록 집중되어있다는 것을 의미하며, 값이 급격하게 변할 수록 선의 가격이 좁게 표시됩니다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'kde')
```

```python
# kind = 'regg' 옵션을 사용할 경우, 회귀선을 확인할 수 있습니다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'regg')
```

```python
# kind = 'hex' 옵션은 kde 속성과 유사하게 밀도를 표현하나 kde처럼 간격이 다르지 않고,
# 간격이 모두 동일한 육각형으로 표현됩니다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'hex')
```

```python
import pandas as pd

raw = pd.read_csv("data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv", encoding = "cp949")
raw.head()
```

```python
# i = 0

for i in range(len(raw)):

    lat = raw.loc[i, "위도"]
    long = raw.loc[i, "경도"]
    name = raw.loc[i, "대피소명칭"]

    print(lat, long, name)
```

```python
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 11)

# 미니맵 추가

minimap = MiniMap()  # M M 대문자
minimap.add_to(m)

# 대피소 마커 추가하기
for i in range(len(raw)):

    lat = raw.loc[i, "위도"]
    long = raw.loc[i, "경도"]
    name = raw.loc[i, "대피소명칭"]

    
    folium.Marker([lat, long],
                 tooltip = name).add_to(m)
    
#     print(lat, long, name)

m
```

```python
m.save('data/Sheltermap.html')
```

clustermarker

```python
from folium.plugins import MarkerCluster
```

```python
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 11)

marker_cluster = MarkerCluster().add_to(m)
# 미니맵 추가

minimap = MiniMap()  # M M 대문자
minimap.add_to(m)

# 대피소 마커 추가하기
for i in range(len(raw)):

    lat = raw.loc[i, "위도"]
    long = raw.loc[i, "경도"]
    name = raw.loc[i, "대피소명칭"]

    
    folium.Marker([lat, long],
                 tooltip = name).add_to(marker_cluster)
    
#     print(lat, long, name)

m
```

```python
m.save('data/Sheltermap.html')
```
